```python
#Import the data
import pandas as pd
```

```python
df1 = pd.read_csv("D:/Shiva Data Science/ExcelR Assignments/Multi Linear Regression/50_Startups.csv", encoding='latin1')
df1
```

```python
#check for missing values
df1.isna().sum()
```

```python
# Standardization X Variable
from sklearn.preprocessing import LabelEncoder
LE = LabelEncoder()
df1["State"] = LE.fit_transform(df1["State"])
df1["State"]
```

```python
# Standardization Y variable
from sklearn.preprocessing import MinMaxScaler
MM = MinMaxScaler()
df1["Profit"] =pd.DataFrame(MM.fit_transform(df1[["Profit"]]))
df1["Profit"]
```

```python
# Correlation matrix
df1.corr()
```

```python
#EDA
# Format the plot background and scatter plots for all the variables
import seaborn as sns
sns.set_style(style="darkgrid")
sns.pairplot(df1)
```

```python
## Step-3 Split the variables as X and Y

#X = df1[["R&D Spend"]]                                                 #============== Case-1
#X = df1[["R&D Spend","Marketing Spend"]]                        #============== Case-2      
X = df1[["R&D Spend","Marketing Spend","Administration"]]             #============== Case-3   <<<Good>>>>
#X = df1[["R&D Spend","Marketing Spend","Administration","State"]]             #============== Case-4 
Y = df1["Profit"]
```

```python
# Model fitting

from sklearn.linear_model import LinearRegression
LR = LinearRegression()
LR.fit(X,Y)
```

```python
# Model predicted values
Y_pred = LR.predict(X)
Y_pred
```

```python
# Model evaluation

from sklearn.metrics import mean_squared_error
mse = mean_squared_error(Y,Y_pred)
print("Mean square error ",mse.round(3))

import numpy as np
print("Root mean square error ", np.sqrt(mse).round(3))
```

```python
from sklearn.metrics import r2_score
r2 = r2_score(Y,Y_pred)
print("r2_score",r2.round(3))
```

```python
## R2 values with cases without model improvement
# Case-1      r2_score 0.947
# Case-2      r2_score 0.950
# Case-3      r2_score 0.951
# Case-4      r2_score 0.951
```

# Improving the model by by Data cleaning

```python
## Built full model
import statsmodels.api as sm

model = sm.OLS(Y, X)  # Capitalize OLS
model = model.fit()
```

```python
#Variance Inflation Factor (VIF) is a measure of multicollinearity

from statsmodels.stats.outliers_influence import variance_inflation_factor

X1 = df1[["R&D Spend","Marketing Spend","Administration","State"]]
vif = [variance_inflation_factor(X1.values, i) for i in range(X1.shape[1])]
vif

## R&D Spend and Marketing Spend are having moderate level of multicollinearity issuess
```

```python
# residual analysis by using Q-Q plot

import matplotlib.pyplot as plt

QQ_plot = sm.qqplot(model.resid,line="q") # line = 45 to draw the diagnoal line
plt.title("Normal Q-Q plot of residuals")
plt.show()

list(np.where(model.resid>10))             # No residuals found
```

```python
# Residual plot for homoscedasticity

model.fittedvalues
model.resid

plt.scatter(model.fittedvalues,model.resid)
plt.title("Residual plot")
plt.xlabel("Fitted values")
plt.ylabel("residual value")
plt.show()
```

```python
#Residual Vs Regressors
fig=plt.figure(figsize=(15,8))
fig=sm.graphics.plot_regress_exog(model,"Administration",fig=fig)
plt.show()
```

```python
fig=plt.figure(figsize=(15,8))
fig=sm.graphics.plot_regress_exog(model,"R&D Spend",fig=fig)
plt.show()
```

```python
fig=plt.figure(figsize=(15,8))
fig=sm.graphics.plot_regress_exog(model,"Marketing Spend",fig=fig)
plt.show()
```

```python
# Model Deletion Diagnostics
# Detecting Influencers/Outliers
# Cookâs Distance

model_influence = model.get_influence()
(cooks , P_value) = model_influence.cooks_distance
(cooks , P_value) 
```

```python
# Plot the influence values using the stemplot

fig = plt.subplots(figsize=(20,7))
plt.stem(np.arange(len(df1)),np.round(cooks,3))
plt.xlabel("Row index")
plt.ylabel("Cooks dis")
plt.show()
```

```python
#index and value of influencer where c is more than 0.5
(np.argmax(cooks),np.max(cooks))
```

```python
# High Influence points

from statsmodels.graphics.regressionplots import influence_plot
influence_plot(model)
plt.show()     # from the plot 38,46 and 49(from cooks distance) are the highly influencers
```

# Improving the Model

```python
# Removing influencer points
df1[df1.index.isin([38,46,49])]
```

```python
df1_new = df1
```

```python
df1_new = df1_new.drop(["Marketing Spend"],axis=1) 
df1_new                   # R&D Spend and Marketing Spend are having moderate level of multicollinearity issues
```

```python
df1_new = df1_new.drop(df1_new.index[[38,46,49]],axis=0).reset_index()
df1_new                   # High influence points
```

```python
X = df1_new[["R&D Spend","Administration"]]             #============== Case-3   <<<Good>>>>
Y = df1_new["Profit"]
```

```python
## Built full model
import statsmodels.api as sm

model2 = sm.OLS(Y, X)  # Capitalize OLS
model2 = model2.fit()
```

```python
# Cookâs Distance

model_influence = model2.get_influence()
(cooks2 , P_value) = model_influence.cooks_distance
(cooks2 , P_value) 

# Plot the influence values using the stemplot

fig = plt.subplots(figsize=(20,7))
plt.stem(np.arange(len(df1_new)),np.round(cooks2,3))
plt.xlabel("Row index")
plt.ylabel("Cooks dis")
plt.show()
```

```python
#index and value of influencer where c is more than 0.5
(np.argmax(cooks2),np.max(cooks2))
```

```python
df1_new2 = df1_new.drop(df1_new.index[[2]],axis=0).reset_index()
df1_new2                   # High influence points
```

```python
# Model evaluation

from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y,y_pred)
print("Mean square error ",mse.round(3))

import numpy as np
print("Root mean square error ", np.sqrt(mse).round(3))
```

```python
from sklearn.metrics import r2_score
r2 = r2_score(y,y_pred)
print("r2_score",r2.round(3))
```
