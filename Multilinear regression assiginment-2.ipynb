```python
#Import the data
import pandas as pd
```

```python
df2 = pd.read_csv("D:/Shiva Data Science/ExcelR Assignments/Multi Linear Regression/ToyotaCorolla.csv", encoding='latin1')
df2
```

```python
#check for missing values
df2.isna().sum()
```

```python
required_columns = ["Price","Age_08_04","KM","HP","cc","Doors","Gears","Quarterly_Tax","Weight"]
df2_1 = df2[required_columns]
df2_1
```

```python
df2_1.shape
```

```python
# Standardization
from sklearn.preprocessing import MinMaxScaler
MM = MinMaxScaler()
df2_1["Price"] =pd.DataFrame(MM.fit_transform(df2_1[["Price"]]))
df2_1["Price"]
```

```python
# Exploratory data analysis
df2_1.corr()
```

```python
# Format the plot background and scatter plots for all the variables
import seaborn as sns
sns.set_style(style="darkgrid")
sns.pairplot(df2_1)
```

```python
# Split the variables X and Y
Y = df2_1["Price"]
#X =df2_1[['Age_08_04']]
#X =df2_1[['Age_08_04','Weight']]
#X =df2_1[['Age_08_04','Weight','KM']]         
X =df2_1[['Age_08_04','Weight','KM','HP']]                     ### Best model fit
#X =df2_1[['Age_08_04','Weight','KM','HP','Quarterly_Tax']]
#X =df2_1[['Age_08_04','Weight','KM','HP','Quarterly_Tax','Doors']]
Y
X
```

```python
# Model fitting

from sklearn.linear_model import LinearRegression
LR = LinearRegression()
LR.fit(X,Y)
```

```python
LR.coef_
LR.intercept_
```

```python
# Model predicted values
Y_pred = LR.predict(X)
Y_pred
```

```python
# Model evaluation

from sklearn.metrics import mean_squared_error
mse = mean_squared_error(Y,Y_pred)
print("Mean square error ",mse.round(3))

import numpy as np
print("Root mean square error ", np.sqrt(mse).round(3))
```

```python
from sklearn.metrics import r2_score
r2 = r2_score(Y,Y_pred)
print("r2_score",r2.round(2))
```

# Improving the model by Data cleaning

```python
## Built full model
import statsmodels.api as sm

model = sm.OLS(Y, X)  # Capitalize OLS
model = model.fit()
```

```python
#Variance Inflation Factor (VIF) is a measure of multicollinearity

from statsmodels.stats.outliers_influence import variance_inflation_factor

X1 = df2_1[['Age_08_04','Weight','KM',"HP"]]  
vif = [variance_inflation_factor(X1.values, i) for i in range(X1.shape[1])]
vif
# Weight and HP having high collinearity
```

```python
# residual analysis by using Q-Q plot

import matplotlib.pyplot as plt

QQ_plot = sm.qqplot(model.resid,line="q") # line = 45 to draw the diagnoal line
plt.title("Normal Q-Q plot of residuals")
plt.show()

list(np.where(model.resid>10))             # No residuals found
```

```python
# Residual plot for homoscedasticity

model.fittedvalues
model.resid

plt.scatter(model.fittedvalues,model.resid)
plt.title("Residual plot")
plt.xlabel("Fitted values")
plt.ylabel("residual value")
plt.show()
```

```python
# Model Deletion Diagnostics
# Detecting Influencers/Outliers
# Cook’s Distance

model_influence = model.get_influence()
(cooks , P_value) = model_influence.cooks_distance
(cooks , P_value) 
```

```python
# Plot the influence values using the stemplot

fig = plt.subplots(figsize=(20,7))
plt.stem(np.arange(len(df2_1)),np.round(cooks,3))
plt.xlabel("Row index")
plt.ylabel("Cooks dis")
plt.show()
```

```python
#index and value of influencer where c is more than 0.5
(np.argmax(cooks),np.max(cooks))
```

```python
# High Influence points

from statsmodels.graphics.regressionplots import influence_plot
influence_plot(model)
plt.show()     
```

```python
from statsmodels.stats.outliers_influence import OLSInfluence
influence = model.get_influence()
threshold = 2.0
influence_summary = influence.summary_frame()
high_influence_points = influence_summary[influence_summary['hat_diag'] > threshold]
high_influence_indices = high_influence_points.index
high_influence_indices          ##There are no high influence INDICES
```

```python
# Removing influencer points
df2_1[df2_1.index.isin([109])]
```

```python
df2_1_new = df2_1
df2_1_new
```

```python
df2_1_new =df2_1_new.drop(["HP"],axis=1)    #Drop HP Having high collinearity
df2_1_new
```

# Predict new model

```python
# Split the variables x and y
y = df2_1_new["Price"]
#x =df2_1_new[['Age_08_04']]
#x =df2_1_new[['Age_08_04','Weight']]
x =df2_1_new[['Age_08_04','Weight','KM']]         ### Best model fit                
y
x
```

```python
# Model fitting

from sklearn.linear_model import LinearRegression
LR = LinearRegression()
LR.fit(x,y)
```

```python
# Model predicted values
y_pred = LR.predict(x)
y_pred
```

```python
# Model evaluation

from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y,y_pred)
print("Mean square error ",mse.round(3))

import numpy as np
print("Root mean square error ", np.sqrt(mse).round(3))
```

```python
from sklearn.metrics import r2_score
r2 = r2_score(y,y_pred)
print("r2_score",r2.round(3))
```
